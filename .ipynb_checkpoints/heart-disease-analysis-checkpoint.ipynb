## 1. Introduction

 Heart disease is one of the leading causes of death globally and is silent and deadly killer. There are no diagnosis until the inviduals experience signs or symptoms of the heart failure, heart attack or arrhythmia. The project goal and motivation are to help the doctors to identify the key factors that contribute to heart disease.

Dataset are obtained from <a href="https://www.kaggle.com/ronitf/heart-disease-uci">this link</a> so I can pay more attention on providing insight and technical part.

## 2. Dataset description

The dataset consists of 303 individuals data where 138 individuals are presented with no heart disease and 165 individuals are presented with heart disease. Initially, there were 76 attributes but 14 attributes/ columns will be used as the dataset, which are described below.

1. **age**: Age in years
2. **sex**: (1 = male, 0 = female)
3. **cp**: Chest pain type (1: typical angina, 2: atypical angina, 3: non-anginal pain, 4: asymptomatic)
4. **trestbps**: Resting blood pressure (in mm Hg on admission to the hospital)
5. **chol**: Serum cholesterol (in mg/dl)
6. **fbs**: Fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
7. **restecg**: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)
8. **thalach**: Maximum heart rate achieved
9. **exang**: exercise-induced angina (1 = yes; 0 = no)
10. **oldpeak**: ST depression induced by exercise relative to rest
11. **slope**: The slope of the peak exercise ST-segment (1: upsloping, 2: flat, 3: downsloping)
12. **ca**: Number of major vessels (0â€“3) colored by fluoroscopy
13. **thal**: Blood disorder called thalassemia (3 = normal, 6 = fixed defect, 7 = reversible defect)
14. **target**: Heart disease (0 = no, 1 = yes)

8. Exercise-induced angina vs Heart Disease

```python
## Creating Stacked barchart function
def stacked_barchart(data, title = None, ylabel = None, xlabel = None):
    default_colors = ['#0333a3', '#c7041f', '#219AD8']
    
    # From raw value to percentage
    totals = data.sum(axis = 1)
    bars = ((data.T / totals) * 100).T
    r = list(range(data.index.size))

    # Plot
    barWidth = 0.95
    names = data.index.tolist()
    bottom = [0] * bars.shape[0]

    # Create bars
    color_index = 0
    plots = []
    for bar in bars.columns:
        plots.append(plt.bar(r, bars[bar], bottom = bottom, color = default_colors[color_index], 
                             edgecolor = 'white', width = barWidth))
        bottom = list(map(add, bottom, bars[bar]))
        color_index = 0 if color_index >= len(default_colors) else color_index + 1

    # Custom x axis
    plt.title(title)
    plt.xticks(r, names)
    plt.xlabel(data.index.name if xlabel is None else xlabel)
    plt.ylabel(data.columns.name if ylabel is None else ylabel)
    ax = plt.gca()
        
    y_labels = ax.get_yticks()
    ax.set_yticklabels([str(y) + '%' for y in y_labels])

    flat_list = [item for sublist in data.T.values for item in sublist]
    for i, d in zip(ax.patches, flat_list):
        data_label = str(d) + " (" + str(round(i.get_height(), 2)) + "%)"
        ax.text(i.get_x() + 0.45, i.get_y() + 5, data_label, horizontalalignment = 'center', verticalalignment = 'center', 
                fontdict = dict(color = 'white', size = 20))

    for item in ([ax.title]):
        item.set_fontsize(27)
        
    for item in ([ax.xaxis.label, ax.yaxis.label] + ax.get_xticklabels() + ax.get_yticklabels()):
        item.set_fontsize(24)
    
    legend = ax.legend(plots, bars.columns.tolist(), fancybox = True)
    plt.setp(legend.get_texts(), fontsize = '20')
```

```python
fig = plt.gcf()
fig.set_size_inches(25, 35)
grid_rows = 3
grid_cols = 2

## Exercise-induced angina vs Heart Disease
plt.subplot(grid_rows, grid_cols, 1)
temp = df[['exercise_induced_angina', 'target']].groupby(['exercise_induced_angina', 'target']).size().unstack('target')
temp.rename(index = {0: 'No', 1: 'Yes'}, columns = {0: 'No Heart Disease', 1: 'Heart Disease'}, inplace = True)
stacked_barchart(temp, title = 'Exercise-induced angina vs Heart Disease', ylabel = 'Percentage')
```

Based on the finding, irrespective of gender there are 69.61% of patients have no exercise-induced angina are likely to get heart disease. However, patients who do not exercise will be higher risk of getting heart disease.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X = df.iloc[:, 0:13]
y = df.iloc[:, -1]

bestfeatures = SelectKBest(score_func = chi2, k = 10)
fit = bestfeatures.fit(X, y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

# Concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores], axis = 1)

# Naming the dataframe columns
featureScores.columns = ['Features', 'Score'] 

# Print all the features
print(featureScores.nlargest(13, 'Score')) 
```

Based on the finding, max_heart_rate_achieved, st_depression, num_major_vessels and chest_pain_type show to be the most important features.

3\. Machine learning

Feature importance property of the model can be used to extract the most important features for the dataset. Feature importance will show the score for each features of dataset, the higher the score the more important or relevant is the feature towards output variable (target). Boruta py will be used to pick important features on random forest.

```python
## Import machine learning models
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import ExtraTreesClassifier
import lightgbm as lgb

## Import features extractor
from boruta import boruta_py

## Import necessary library
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, cohen_kappa_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix
import itertools

X = df.iloc[:, 0:13].values
y = df.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)

# pipe = Pipeline([('scaler', StandardScaler()), ('rf', RandomForestClassifier())])
# pipe.fit(X_train, y_train)
# print(pipe.score(X_test, y_test))

## Standardize the data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
## Feature extractor function
def feature_imp(columns, feature_importances):
    fi = pd.DataFrame()
    fi["feature"] = columns
    fi["importance"] = feature_importances
    return fi.sort_values(by="importance", ascending = False)
```

1. Random Forest

```python
## Random Forest
random_f = RandomForestClassifier(n_estimators = 10, random_state = 42)

random_f.fit(X_train, y_train)

# Accurancy Score of Train and Test
rf_train_score = accuracy_score(y_train, random_f.predict(X_train)) * 100
rf_test_score = accuracy_score(y_test, random_f.predict(X_test)) * 100

# Prediction
rf_prediction = random_f.predict(X_test)

# Recall Score
rf_recallscore = recall_score(y_test, rf_prediction)

# Precision score
rf_precision = precision_score(y_test, rf_prediction)

# F1 score
rf_f1score = f1_score(y_test, rf_prediction) 

# Features importance of random forest using boruta py
random_forest = RandomForestClassifier()
feature_selector = boruta_py.BorutaPy(random_forest, n_estimators = 'auto', verbose = 2)

feature_selector.fit(X_train, y_train)

rf_most_important_features = df.columns[:-1][feature_selector.support_].tolist()
rf_most_important_features = rf_most_important_features[0:6]
```

```python
# Plotting of features importance of Random Forest
feature_imp(df.iloc[:, 0:13].columns, random_f.feature_importances_).plot('feature', 'importance', 'barh', figsize = (12, 7), legend = False)
plt.title('Important Features by Random Forest', fontsize = 20)
plt.ioff()
```

2. Decision Tree

```python
## Decision Tree
decision_tree = DecisionTreeClassifier()

decision_tree.fit(X_train, y_train)

dt_train_score = accuracy_score(y_train, decision_tree.predict(X_train)) * 100
dt_test_score = accuracy_score(y_test, decision_tree.predict(X_test)) * 100

# Prediction
dt_prediction = decision_tree.predict(X_test)

# Recall Score
dt_recallscore = recall_score(y_test, dt_prediction)

# Precision score
dt_precision = precision_score(y_test, dt_prediction)

# F1 score
dt_f1score = f1_score(y_test, dt_prediction) 

## DT most important features
feature_importance = pd.DataFrame()
feature_importance["feature"] = df.iloc[:, 0:13].columns
feature_importance["importance"] = decision_tree.feature_importances_

dt_most_important_features = feature_importance.sort_values(by = 'importance', ascending = False)
dt_most_important_features = dt_most_important_features.iloc[0:6]['feature'].to_list()
```

```python
# Plotting of features importance of Decision Tree
feature_imp(df.iloc[:, 0:13].columns, decision_tree.feature_importances_).plot('feature', 'importance', 'barh', figsize = (12, 7), legend = False)
plt.title('Important Features by Decision Tree', fontsize = 20)
plt.ioff()
```

All Model Performance Metrics (Accuracy Score, Recall Score, Precision, F1 Score)

```python
## Show model performance report function
def model_report(accuracy, recallscore, precision, f1score, name):
    df = pd.DataFrame({"Model"           : [name],
                       "Accuracy Score"  : [accuracy],
                       "Recall Score"    : [recallscore],
                       "Precision"       : [precision],
                       "F1 Score"        : [f1score]
                      })
    return df
```

```python
## Outputs for every model
model1 = model_report(rf_test_score, rf_recallscore, rf_precision, rf_f1score, "Random Forest")
model2 = model_report(dt_test_score, dt_recallscore, dt_precision, dt_f1score, "Decision Tree")
model3 = model_report(et_test_score, et_recallscore, et_precision, et_f1score, "Extra Tree")
model4 = model_report(lr_test_score, log_reg_recallscore, log_reg_precision, log_reg_f1score, "Logistic Regression")
model5 = model_report(xgb_test_score, xgb_recallscore, xgb_precision, xgb_f1score, "XGBoost Classifier")
model6 = model_report(svm_test_score, svm_recallscore, svm_precision, svm_f1score, "Support Vector Machine")
model7 = model_report(knn_test_score, knn_recallscore, knn_precision, knn_f1score, "K-nearest neighbors")
model8 = model_report(gnb_test_score, gnb_recallscore, gnb_precision, gnb_f1score, "Naive Bayes")
model9 = model_report(lgbm_test_score, lgbm_recallscore, lgbm_precision, lgbm_f1score, "LightGBM")

model_performances = pd.concat([model1, model2, model3, model4, model5, 
                                model6, model7, model8, model9], axis = 0).reset_index()
model_performances = model_performances.drop(columns = "index", axis =1)
model_performances.index += 1
model_performances
```

Recall is important than precision because cost of false negative is high and false positive is not as much. Because patients will die if they have heart disease but the doctors ask you don\`t get the treatment. All recall score of the machine learning models is higher than precision. These machine learning models are doing good because false negative is costly.
